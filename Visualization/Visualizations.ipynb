```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

Information_Mart = pd.read_csv("../Information_Mart/orders_products_details.csv")

# Histogram of list price
plt.figure(figsize=(10, 6))
sns.histplot(Information_Mart['list_price_egp'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of List Price')
plt.xlabel('List Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of list price by category
plt.figure(figsize=(10, 6))
sns.boxplot(data=Information_Mart, x='category_id_product', y='list_price_egp', palette='pastel')
plt.title('List Price by Category')
plt.xlabel('Category ID')
plt.ylabel('List Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=Information_Mart, x='status_description', palette='pastel')
plt.title('Bar Plot of Order Status Counts')
plt.xlabel('Order Status')
plt.ylabel('Count')
plt.show()
```
